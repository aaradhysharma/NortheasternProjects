```python

iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target
Z = 5
linear_svc = svm.SVC(kernel='linear', degree=3, C=Z).fit(X, y)
rbf_svc = svm.SVC(kernel='rbf',degree=3, C=Z).fit(X, y)
poly_svc = svm.SVC(kernel='poly', degree=3, C=Z).fit(X, y)
```
